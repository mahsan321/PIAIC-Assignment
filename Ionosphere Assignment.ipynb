```python
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras import losses
from tensorflow.keras import metrics

from sklearn.model_selection import train_test_split 
import tensorflow as tf
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\Muhammad Ahsan\AI-Q2-learning-resources\DLAssignments\ionosphere_data.csv')
df.head()
print(df.columns)
```

```python
df.fillna(df.mean())
```

```python
#shuffle data
df = df.sample(frac=1)
df.head()
```

```python
y = df["label"]
dfn = df.drop(['label', 'feature2'], axis = 1)
for c in range(len(y)):
    if y.iloc[c] == 'g':
        y.iloc[c] = 1
    else:
        y.iloc[c] = 0
dfn.head()
```

```python
dfn = dfn.astype("float64")
```

```python
# Applying Train Test Split
X_train, X_test, y_train, y_test = train_test_split(dfn, y, test_size=0.4, random_state=42)
#X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.29, random_state=42)
```

```python
mean = X_train.mean(axis=0)

std = X_train.std(axis=0)

X_train -= mean
X_train /= std
X_test -= mean
X_test /= std
```

```python
X_train.head()
```

```python
X_train.shape[1]
y_train = tf.convert_to_tensor(y_train, dtype=tf.float32)
y_test = tf.convert_to_tensor(y_test, dtype=tf.float32)
```

```python
def build_model():
    model = models.Sequential()
    model.add(layers.Dense(1, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(layers.Dense(16, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(optimizer=optimizers.RMSprop(lr=0.001), loss=losses.binary_crossentropy, metrics=[metrics.binary_accuracy])
    return model
```

```python
mdl = build_model()
```

```python
histrory = mdl.fit(X_train, y_train, epochs=250, verbose=2)
```

```python
test_loss, test_acc = mdl.evaluate(X_test, y_test)
print('test_acc:', test_acc)
```
